##### TRYB PERMISSIVE

```
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

# Definiowanie schematu
schema = StructType([
    StructField("id", IntegerType(), True),    
    StructField("name", StringType(), True), 
    StructField("age", IntegerType(), True)   
])

#TRYB PERMISSIVE
df_permissive = spark.read \
    .schema(schema) \
    .format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("mode", "PERMISSIVE") \
    .load("dbfs:/FileStore/shared_uploads/data_with_errors.csv")

# Wyświetlenie danych
print("Tryb PERMISSIVE:")
display(df_permissive)
```

##### TRYB DROPMALFORMED

```
#TRYB DROPMALFORMED
df_drop_malformed = spark.read \
    .schema(schema) \
    .format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("mode", "DROPMALFORMED") \
    .load("dbfs:/FileStore/shared_uploads/data_with_errors.csv")

# Wyświetlenie danych
print("Tryb DROPMALFORMED:")
display(df_drop_malformed)
#wiersze z nieprawidłowymi danymi w kolumnie age zostały pominięte 
```

##### TRYB FAILFAST

```
#TRYB FAILFAST
df_fail_fast = spark.read \
        .schema(schema) \
        .format("csv") \
        .option("header", "true") \
        .option("mode", "FAILFAST") \
        .load("dbfs:/FileStore/shared_uploads/data_with_errors.csv")


print("Tryb FAILFAST:")
display(df_fail_fast)
```

### ZADANIE 4
##### Użycie DataFrameWriter.
Zapisz jeden z wybranych plików do formatów (‘.parquet’, ‘.json’). Sprawdź, czy dane są zapisane poprawnie, użyj do tego spark.read (DataFrameReader). 


```
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

# Definiowanie schematu
schema = StructType([
    StructField("id", IntegerType(), True),    # Kolumna "id" typu Integer
    StructField("name", StringType(), True),   # Kolumna "name" typu String
    StructField("age", IntegerType(), True)    # Kolumna "age" typu Integer
])
df = spark.read \
    .schema(schema) \
    .format("csv") \
    .option("header", "true") \
    .load("dbfs:/FileStore/shared_uploads/data_with_errors.csv")

# Wyświetlenie danych
print("Dane wczytane z pliku CSV:")
display(df)


# Zapisywanie danych do formatu Parquet
parquet_path = "dbfs:/FileStore/shared_uploads/data_with_errors.parquet"
df.write \
    .format("parquet") \
    .mode("overwrite") \
    .save(parquet_path)


# Zapisywanie danych do formatu JSON
json_path = "dbfs:/FileStore/shared_uploads/data_with_errors.json"
df.write \
    .format("json") \
    .mode("overwrite") \
    .save(json_path)

# Wczytanie danych z formatu Parquet
df_parquet = spark.read \
    .format("parquet") \
    .load(parquet_path)

print("Dane wczytane z pliku Parquet:")
display(df_parquet)

# Wczytanie danych z formatu JSON
df_json = spark.read \
    .schema(schema) \
    .format("json") \
    .load(json_path)

print("Dane wczytane z pliku JSON:")
display(df_json)

```
